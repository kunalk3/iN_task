**Question 1**

Write a Python Program to implement your own myreduce() function which works exactly like Python's built-in function reduce()

```
def myreduce(func, values):
  result = values[0]
  for i in values[1:]:
    result = func(result, i)
  return result

def func(a, b):
  return a * b

l = [2, 5, 6, 1, 8, 3, 10]
print(f"Output of the function myreduce(): {myreduce(func, l)}")
```

**Question 2**

Write a Python program to implement your own myfilter() function which works exactly like Python's built-in function filter()

```
def myfilter(fun, values):
  result = list()
  for i in values:
    if fun(i):
      result.append(i)
  return result
  
def fun(x):
  if x % 5 == 0:
    return True
  else:
    return False

lst = [12, 15, 60, 34, 53]
print(f"My Filter Output is: {myfilter(fun, lst)}")
```

**Question 3**

Implement List comprehensions to produce the following lists.

1. ['x', 'xx', 'xxx', 'xxxx', 'y', 'yy', 'yyy', 'yyyy', 'z', 'zz', 'zzz', 'zzzz']

2. ['x', 'y', 'z', 'xx', 'yy', 'zz', 'xxx', 'yyy', 'zzz', 'xxxx', 'yyyy', 'zzzz']

3. [[2], [3], [4], [3], [4], [5], [4], [5], [6]]

4. [[2, 3, 4, 5], [3, 4, 5, 6], [4, 5, 6, 7], [5, 6, 7, 8]]

5. [(1, 1), (2, 1), (3, 1), (1, 2), (2, 2), (3, 2), (1, 3), (2, 3), (3, 3)]

```
l1 = ['x', 'y', 'z']
print([i*j for j in l1 for i in range(1, 5)], '\n')

l2 = ['x', 'y', 'z']
print([i*j for j in range(1, 5) for i in l2], '\n')

print([[x + y] for y in range(3) for x in range(2, 5)], '\n')

print([[a+b for b in range(2, 6)] for a in range(4)], '\n')

print([(p, q) for q in range(1, 4) for p in range(1, 4)])
```
