*area = (s*(s-a)*(s-b)*(s-c)) ** 0.5*

```
class area:
    def __init__(self,a,b,c):
        self.a = float(a)
        self.b = float(b)
        self.c = float(c)
        
a= int(input("a="))
b= int(input("b="))
c= int(input("c="))

class triangle(area):
    def __init__(self,a,b,c):
        super().__init__(a,b,c)

    def get_area(s3elf):
        s = (a + b + c) / 2
        return (s*(s-a)*(s-b)*(s-c)) ** 0.5        

t = triangle(a,b,c)
print(f"area  of triangle with {a},{b},{c} is = ", t.get_area())
```

**Question 2**

Write a function filter_long_words() that takes a list of words and an integer n and returns 2the list of words that are longer than n.

```
def filter_long_words(List,Integer):
    list_of_words = []
 
    for i in range(len(List)):
        if len(List[i]) > Integer:
            list_of_words.append(List[i])

    return list_of_words

print(filter_long_words(['laptop','laptopCharger','laptopBag','laptops','laptopFan'],7))
```

**Question 3**

Write a Python program using function concept that maps list of words into a list of integers representing the lengths of the corresponding words

```
def map_Words_to_Length(List):
    return list(map(len, List))

word_List=list(input("Enter words : ").split(","))
List=[x.strip() for x in word_List]
Words_lengths=map_Words_to_Length(List)

print("Length of Words are :",Words_lengths )
```
